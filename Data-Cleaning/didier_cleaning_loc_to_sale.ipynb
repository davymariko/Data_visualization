# Change Data type

| Syntax | action | Description |
| ----------- | ----------- | ----------- |
| Header | action | Title |
| locality             | remains  | string |==> some in postcode format , others in city name
| postcode             | converts | into int64 |   
| house_is             | remains  | string |
| proprety_subtype     | remains  | string |

| sale                 | remains  | string | --> change empty str --> Unknown, None --> Unkown, 

#### 1. PostCode

```python
# 1 Converts postCode into int64

loc_to_sales_df['postcode'] = loc_to_sales_df['postcode'].astype('Int64')
loc_to_sales_df['postcode'].dtypes
```

#### 2. Price

```python
# 2 Converting price
def grabs_strips(x):
    if type(x) == str:
        # return x.str.extract('(\d*\.?\d*)', expand=False).astype(float)
        return re.match(r'\d*\.?\d*',x).group()
    return x


loc_to_sales_df['price'] = loc_to_sales_df['price'].apply(grabs_strips)
```

```python
print(loc_to_sales_df['price'].dtypes)
loc_to_sales_df['price']
```

```python
for x in loc_to_sales_df['price']:
  
    if type(x) is not str:
        print(x, type(x))
```

##### checking if number of NaN == number of non string

```python
count = 0
for x in loc_to_sales_df['price']:
  
    if type(x) is not str:
        count += 1

# How much NaN values in % if less than 60% then later convert then in mean
print(count == loc_to_sales_df['price'].isna().sum())
count / loc_to_sales_df['price'].count() * 100 # less than 60 % maybe we could replace by all nan by mean
```

```python
# Conversion into float
loc_to_sales_df['price'] = pd.to_numeric(loc_to_sales_df['price'], errors='coerce')
```

```python
# Checking if price is float
loc_to_sales_df['price']
```

#### Replace all NaN by mean --> no downgrading total calculations

```python
loc_to_sales_df['price'].fillna(loc_to_sales_df['price'].mean()).round(2)
```
